---
## 1 — ALeRCE and the ZTF Alert Stream

[ALeRCE](https://alerce.science/) (Automatic Learning for the Rapid Classification of Events) is a Chilean-led astronomical broker that processes the real-time alert stream from the **Zwicky Transient Facility (ZTF)**. ZTF scans the entire northern sky every few nights and generates an alert every time a brightness change is detected somewhere on the sky. ALeRCE ingests, stores, and classifies these alerts, making them available through a public API.

### What is a ZTF alert?

Each alert contains two complementary representations of the event:

- **Image stamps** — three 21×21 pixel cutouts centred on the source: the *science* image (current observation), the *reference* image (archival baseline), and their *difference*. These encode the morphology and context of the transient at the moment of detection.

- **Light curve** — the full time series of brightness measurements (`magpsf` ± `sigmapsf`) across both ZTF bands (*g* = band 1, *r* = band 2) for every detection linked to that sky position.

### Labels: classifiers as a source of ground truth

ALeRCE runs two machine-learning classifiers whose predictions we use as class labels:

1. **Stamp classifier** ([Carrasco-Davis et al. 2021](https://arxiv.org/abs/2008.03309)) — classifies each object from the very first alert stamp into five broad categories: **AGN**, **SN**, **VS** (variable star), **Asteroid**, **Bogus**. We use only objects with classification probability > 0.8 to keep the labels clean.

2. **Light curve classifier** ([Sánchez-Sáez et al. 2021](https://arxiv.org/abs/2008.03311)) — classifies objects from their full light curve into 15 sub-classes. We focus on the four **transient** sub-classes: **SNIa**, **SNIbc**, **SNII**, **SLSN**. Again, we keep only objects with probability > 0.8.

```python
import os
import pickle
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

from alerce.core import Alerce

client = Alerce()

DATASET_DIR = Path("dataset")
DATASET_DIR.mkdir(exist_ok=True)
```

---
## 3 — Image Stamp Dataset

### 3.1 — What we are collecting

For each object we retrieve the triplet of 21×21 pixel cutouts from its **first detection**: science, reference and difference images. Stacked along the channel dimension, each sample is a tensor of shape **(3, 21, 21)** — the input format our Vision Transformer will expect.

We query 100 objects per class from the stamp classifier (probability > 0.8), for a total of **500 labelled samples**.

![Stamp examples](figures/class_samples.png)

```python
STAMP_CLASSES = ["AGN", "SN", "VS", "asteroid", "bogus"]
N_OBJECTS_PER_CLASS = 100
MIN_PROBABILITY = 0.8


def query_stamp_objects(class_name: str, n_objects: int, min_prob: float) -> pd.DataFrame:
    """Query objects classified by the stamp classifier with high confidence."""
    df = client.query_objects(
        classifier="stamp_classifier",
        class_name=class_name,
        probability=min_prob,
        ranking=1,
        order_by="probability",
        order_mode="DESC",
        count=False,
        page_size=n_objects,
        format="pandas",
    )
    df = df.set_index("oid")
    print(f"  {class_name}: {len(df)} objects retrieved")
    return df


stamp_objects = {}
for cl in STAMP_CLASSES:
    stamp_objects[cl] = query_stamp_objects(cl, N_OBJECTS_PER_CLASS, MIN_PROBABILITY)
```

### 3.2 — Download stamps

For each object we fetch the detection list, locate the first alert that has a stamp stored, and download the triplet. Failed downloads are skipped and logged.

```python
def download_stamp(oid: str) -> np.ndarray | None:
    """Download the first-detection stamp triplet for a given object.

    Returns an array of shape (3, 21, 21) — [science, reference, difference] —
    or None if the download fails.
    """
    try:
        detections = client.query_detections(oid, format="pandas")
        detections = detections.sort_values("mjd")
        stamped = detections.loc[detections["has_stamp"]]
        if stamped.empty:
            return None
        candid = stamped.iloc[0]["candid"]
        raw = client.get_stamps(oid, candid)
        science, reference, difference = raw[0].data, raw[1].data, raw[2].data
        stamp = np.stack([science, reference, difference], axis=0)  # (3, 21, 21)
        return stamp.astype(np.float32)
    except Exception as exc:
        print(f"    [SKIP] {oid}: {exc}")
        return None


stamps_list, labels_list, oids_list = [], [], []

for label_idx, (cl, df) in enumerate(stamp_objects.items()):
    print(f"Downloading stamps for class '{cl}' ({len(df)} objects)...")
    for oid in df.index:
        stamp = download_stamp(oid)
        if stamp is not None:
            stamps_list.append(stamp)
            labels_list.append(label_idx)
            oids_list.append(oid)

stamps_array = np.stack(stamps_list, axis=0)   # (N, 3, 21, 21)
labels_array = np.array(labels_list)            # (N,)
oids_array   = np.array(oids_list)              # (N,)

print(f"\nStamp dataset shape: {stamps_array.shape}")
print(f"Label distribution: { {cl: int((labels_array == i).sum()) for i, cl in enumerate(STAMP_CLASSES)} }")
```

### 3.3 — Visualise examples

```python
fig, axes = plt.subplots(
    len(STAMP_CLASSES), 3,
    figsize=(7, len(STAMP_CLASSES) * 2.2),
    gridspec_kw={"wspace": 0.05, "hspace": 0.3},
)
channel_names = ["Science", "Reference", "Difference"]

for row, (cl, label_idx) in enumerate(zip(STAMP_CLASSES, range(len(STAMP_CLASSES)))):
    # Pick the first available sample for this class
    idx = np.where(labels_array == label_idx)[0][0]
    stamp = stamps_array[idx]  # (3, 21, 21)
    for col in range(3):
        ax = axes[row, col]
        ax.imshow(np.arcsinh(stamp[col]), cmap="viridis", origin="lower")
        ax.axis("off")
        if row == 0:
            ax.set_title(channel_names[col], fontsize=9)
        if col == 0:
            ax.set_ylabel(cl, fontsize=9, rotation=0, labelpad=40, va="center")

fig.suptitle("One stamp per class (science | reference | difference)", fontsize=11, y=1.01)
plt.savefig(DATASET_DIR / "stamp_examples.png", bbox_inches="tight", dpi=120)
plt.show()
```

### 3.4 — Save to disk

```python
np.savez(
    DATASET_DIR / "stamps.npz",
    stamps=stamps_array,       # float32 (N, 3, 21, 21)
    labels=labels_array,       # int64   (N,)
    oids=oids_array,           # str     (N,)
    class_names=np.array(STAMP_CLASSES),
)
print("Saved:", DATASET_DIR / "stamps.npz")
```

## 4 — Light Curve Dataset

### 4.1 — What we are collecting

A light curve is an **irregularly sampled time series** of brightness measurements. Each data point consists of:
- `mjd` — observation time (Modified Julian Date)
- `magpsf` — PSF-fit magnitude
- `sigmapsf` — photometric uncertainty
- `fid` — filter band (1 = *g*, 2 = *r*)

We restrict ourselves to the four **transient** classes from the light curve classifier — SNIa, SNIbc, SNII, SLSN — which are the scientifically most interesting for time-domain astronomy. We query 100 objects per class with probability > 0.8, for a total of up to **400 labelled light curves**.

The key challenge for the transformer will be handling sequences of **variable length** across different objects.

```python
LC_CLASSES = ["SNIa", "SNIbc", "SNII", "SLSN"]


def query_lc_objects(class_name: str, n_objects: int, min_prob: float) -> pd.DataFrame:
    """Query transient objects classified by the light curve classifier."""
    df = client.query_objects(
        classifier="lc_classifier",
        class_name=class_name,
        probability=min_prob,
        ranking=1,
        order_by="probability",
        order_mode="DESC",
        count=False,
        page_size=n_objects,
        format="pandas",
    )
    df = df.set_index("oid")
    print(f"  {class_name}: {len(df)} objects retrieved")
    return df


lc_objects = {}
for cl in LC_CLASSES:
    lc_objects[cl] = query_lc_objects(cl, N_OBJECTS_PER_CLASS, MIN_PROBABILITY)
```

### 4.2 — Download detections

```python
DETECTION_COLS = ["mjd", "magpsf", "sigmapsf", "fid"]


def download_lightcurve(oid: str) -> pd.DataFrame | None:
    """Download and clean detections for a single object.

    Returns a DataFrame with columns [mjd, magpsf, sigmapsf, fid] sorted by mjd,
    or None if the download fails or the object has no usable detections.
    """
    try:
        det = client.query_detections(oid, format="pandas")
        det = det[DETECTION_COLS].dropna().sort_values("mjd").reset_index(drop=True)
        if len(det) < 5:
            return None
        return det
    except Exception as exc:
        print(f"    [SKIP] {oid}: {exc}")
        return None


lc_records = []  # list of dicts: {oid, class, detections_df}

for cl, df in lc_objects.items():
    print(f"Downloading light curves for class '{cl}' ({len(df)} objects)...")
    for oid in df.index:
        det = download_lightcurve(oid)
        if det is not None:
            lc_records.append({"oid": oid, "label": cl, "detections": det})

print(f"\nTotal light curves collected: {len(lc_records)}")
print("Per class:", {cl: sum(1 for r in lc_records if r["label"] == cl) for cl in LC_CLASSES})
```

### 4.3 — Visualise examples

```python
COLORS = {1: "#56E03A", 2: "#D42F4B"}
BAND_LABELS = {1: "g", 2: "r"}

fig, axes = plt.subplots(2, 2, figsize=(14, 8))
axes = axes.flatten()

# One representative light curve per class
for ax, cl in zip(axes, LC_CLASSES):
    record = next(r for r in lc_records if r["label"] == cl)
    det = record["detections"]
    for fid in [1, 2]:
        mask = det["fid"] == fid
        if mask.sum() == 0:
            continue
        ax.errorbar(
            det.loc[mask, "mjd"],
            det.loc[mask, "magpsf"],
            yerr=det.loc[mask, "sigmapsf"],
            fmt="o", ms=3, color=COLORS[fid], label=BAND_LABELS[fid],
        )
    ax.invert_yaxis()
    ax.set_title(f"{cl} — {record['oid']}", fontsize=10)
    ax.set_xlabel("MJD")
    ax.set_ylabel("Magnitude")
    ax.legend(fontsize=8)

fig.suptitle("Example light curves per transient class", fontsize=12)
plt.tight_layout()
plt.savefig(DATASET_DIR / "lc_examples.png", dpi=120)
plt.show()
```

### 4.4 — Save to disk

Light curves have variable lengths, so we save them in two complementary files:
- **`lc_detections.parquet`** — long-format table with all observations (one row per detection), with `oid` as a grouping key.
- **`lc_metadata.parquet`** — one row per object with its `oid` and `label`.

```python
# Build long-format detections table
det_frames = []
for record in lc_records:
    frame = record["detections"].copy()
    frame["oid"] = record["oid"]
    frame["label"] = record["label"]
    det_frames.append(frame)

lc_detections = pd.concat(det_frames, ignore_index=True)
lc_metadata   = pd.DataFrame([{"oid": r["oid"], "label": r["label"]} for r in lc_records])

lc_detections.to_parquet(DATASET_DIR / "lc_detections.parquet", index=False)
lc_metadata.to_parquet(DATASET_DIR / "lc_metadata.parquet",   index=False)

print("Saved:")
print(f"  {DATASET_DIR}/lc_detections.parquet  — {len(lc_detections):,} rows")
print(f"  {DATASET_DIR}/lc_metadata.parquet    — {len(lc_metadata):,} objects")
```
